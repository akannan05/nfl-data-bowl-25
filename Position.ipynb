```python
import numpy as np
import pandas as pd

import extract
```

```python
extractlist = extract.extract(2022091200, 109)
extractlist
```

```python
defense = extract.separate(extractlist)[0]
weapons = extract.separate(extractlist)[1]
```

```python
print(defense[0].head())
print(defense[1].head())
```

```python
pos_defense = []
for d in defense:
    pos_defense.append(d[['name', 'player_x', 'player_y']])
    
dist_each_def = []

for pos in pos_defense:
    new_df = pd.DataFrame({'euc_position': [list(xy) for xy in zip(pos['player_x'], pos['player_y'])]})
    new_df['name'] = pos.at[1, 'name']
    new_df = new_df[['name', 'euc_position']]
    dist_each_def.append(new_df)
        
print(dist_each_def)
```

```python
pos_offense = []
for w in weapons:
    pos_offense.append(w[['name', 'player_x', 'player_y']])
    
dist_each_weapon = []
for pos in pos_offense:
    new_df = pd.DataFrame({'euc_position': [list(xy) for xy in zip(pos['player_x'], pos['player_y'])]})
    new_df['name'] = pos.at[1, 'name']
    new_df = new_df[['name', 'euc_position']]
    dist_each_weapon.append(new_df)
    
print(dist_each_weapon)
```

```python
import extract
```

```python
for d_d in dist_each_def:
    for i, d_w in enumerate(dist_each_weapon):
        euc_dist = extract.euc_distance(d_d['euc_position'], d_w['euc_position'])
        d_d[f'distance to {i}'] = euc_dist
        
print(dist_each_def)
```

```python
import math
```

```python
def euc_distance_test(pos1, pos2):
    dist = []
    for p in range(len(pos1)):
        dist.append(math.sqrt((pos1[0] - pos2[0])**2 + (pos1[1] - pos2[1])**2))
    return dist
```

```python
for d_d in dist_each_def:
    for i, d_w in enumerate(dist_each_weapon):
        euc_dist = euc_distance_test(d_d['euc_position'], d_w['euc_position'])
        d_d[f'distance to {i}'] = euc_dist
        
print(dist_each_def)
```
